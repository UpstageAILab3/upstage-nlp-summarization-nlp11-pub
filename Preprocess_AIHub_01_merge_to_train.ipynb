### AI Hub 한국어 대화 요약 데이터셋을 원본 train.csv 파일에 붙이기.

```python
import pandas as pd

df_orig = pd.read_csv('../../data/train_orig.csv')  # 수정 없는 원본 train.csv 파일.
df_AIHub_01 = pd.read_csv('../../data/train_AIHub_01.csv')
```

```python
df_orig
```

```python
df_AIHub_01
```

```python
result = pd.concat([df_orig, df_AIHub_01], axis=0, ignore_index=True)
result
```

```python
result.to_csv('../../data/train.csv', encoding='utf-8-sig', index=False)
```
